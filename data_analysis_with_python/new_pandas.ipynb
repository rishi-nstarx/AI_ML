## Pandas – DataFrame and Series

Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning.
It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object,
while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with
labeled axes (rows and columns).

```python
# Series - 1D (similar to numpy 1D array.)
# Dataframe - 2D

# Series(Similar to column in a table.)
import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
print(type(series))

code = ['a', 'b', 'c', 'd', 'e']
series = pd.Series(code)
print(series)
print(type(series))


# Create a series from dictionary.
# output = zip(code, data) # Zip aggregates elements from multiple iterables (like lists, tuples, or strings) into a single iterable of tuples.
dict_data = dict(zip(code, data))
print(dict_data)
dict_series = pd.Series(dict_data) # here we have given our own indexs for the series that is in the form of dict.
print(dict_series)


# another way.
dict_series = pd.Series(data, index=code) # passing separate list(sequences).
print(dict_series)

#-----------------------------------------------------------------------------------------------

# Dataframe (Similar to DB table or excel.)

# Create DF(Dateframe) with the dict of list(value will be the list).
data={
    'Name':['Krish', 'Jhon', 'Jack'],
    'Age':[25,30,35],
    'City':['Banglore', 'New York', 'Florida']
}

df = pd.DataFrame(data)
print(df)
print(type(df))

# import numpy as np
# arr = np.array(df)
# print(arr)

# Create DF(Dateframe) with the list of dict.
data = [
    {'Name': 'Krish', 'Age':25, 'City':'Banglore'},
    {'Name': 'Jhon', 'Age':30, 'City':'New York'},
    {'Name': 'Jack', 'Age':35, 'City':'Florida'}
]

df = pd.DataFrame(data)
print(df)
print(df["Name"])
print(type(df["Name"])) # <class 'pandas.core.series.Series'>
print(type(df))
print(df.loc[1])   # Access row with index label 1
print(df.iloc[1])  # Access row at position 1 (second row)
print(df.iloc[1][1])
print(df.at[2, "Age"])
# Accessing specified element using iat.
print(df.iat[2,2])

# Data manipulation with Dataframe.
df["Salary"] = [10000, 20000, 30000] # Adding a column.
print(df)
# df = df.drop('Salary', axis=1) # axix=0(row)(default) axix=1(column) [It doesn't actually drop the any thing insted returns new table.]
# print(df)
df.drop('Salary', axis=1, inplace=True) # 'inplace=True' Modifies the original DataFrame.
print(df)
df["Age"]=df["Age"]+1
print(df)


df = pd.read_csv("../data/data.csv")
print(df)
print(df.head(5))
print(type(df))


# Display the data types of each column
print("Data types:\n", df.dtypes)

# Describe the DataFrame
print("Statistical summary:\n", df.describe())

# Group by a column and perform an aggregation
grouped = df.groupby('Pulse')['Calories'].mean()
print("Mean Calories by Pulse:\n", grouped)
```
