```python
corpus="""Hello Welcome,to Krish Naik's NLP Tutorials.
Please do watch the entire course! to become expert in NLP.
"""
```

```python
print(corpus)
```

```python
##  Tokenization
## paragraphs-->sentence
from nltk.tokenize import sent_tokenize

# sent_tokenize:- Tokenizes into sentences wrt ".", "!" and etc.
```

```python
documents=sent_tokenize(corpus)
print(documents)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Tokenization 
## Paragraph-->words
## sentence--->words
from nltk.tokenize import word_tokenize

# word_tokenize:- Tokenizes into words.
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
# wordpunct_tokenize:- this is sharper version of "word_tokenize". it takes care of all the punctuation marks.
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer

# Slightly different than "wordpunct_tokenize" find the difference with the o/p or on internet.
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
